```python

ratings = pd.read_csv(r'C:\Users\Nithya\Documents\ML\ml-100k\u.data',sep="\t",usecols=['userId','itemId','rating'])
ratings.head()
```

```python
row=max(ratings.userId)
row
```

```python
col=max(ratings.itemId)
col
```

```python
arr=np.zeros([row,col],dtype=int)
```

```python
print(arr)
```

```python
for i in range(0,100000):
    arr[ratings.userId[i]-1][ratings.itemId[i]-1]=ratings.rating[i]
```

```python
print(arr)
```

```python
euclArr = np.zeros([row,row],dtype=object)
```

```python
for i in range(0,row):
    for j in range(0,row):
        dist = np.linalg.norm(arr[i]-arr[j])
        euclArr[i][j] = (dist,j)
```

```python
for i in range(0,row):
    euclArr[i].sort(axis=0)
```

```python
euclArr
```

```python
k_val = eval(input("value of k : "))
```

```python
tmpList = []
for i in range(0,k_val):
    tmpList+=[euclArr[0][i]]
```

```python
tmpList
```

```python
tmpList[1][1]

```

```python
movRating=0
c = 0
for i in range(0,col):
    if(arr[0][i] == 0):
        for j in range(0,k_val):
            if(arr[tmpList[j][1]][i] !=0):
                c+=1
                movRating+=arr[tmpList[j][1]][i]
        if(c!=0 and round(movRating/c,2)>=2.5):
            print("Movie : ",i,"predicted rating : ",round(movRating/c,2))
        movRating=0
        c=0
            
```

```python
# for specific user
userInput = eval(input("for a specific user 0-942: "))
tmpList = []
for i in range(userInput,k_val):
    tmpList+=[euclArr[userInput][i]]

movRating=0
c = 0
for i in range(0,col):
    if(arr[userInput][i] == 0):
        for j in range(0,k_val):
            if(arr[tmpList[j][1]][i] !=0):
                c+=1
                movRating+=arr[tmpList[j][1]][i]
        if(c!=0 and round(movRating/c,2)>=2.5):
            print("Movie : ",i,"predicted rating : ",round(movRating/c,2))
        movRating=0
        c=0
```
